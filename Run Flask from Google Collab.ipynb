```python
from flask_ngrok import run_with_ngrok
```

```python
from flask import Flask
```

```python
app = Flask(__name__)
```

```python
run_with_ngrok(app)
@app.route("/")
def index():
  return "<h1>Trail of Flask on Google Collab!</h1>"
@app.route("/get_details")
def get_details():
  return "<h1>This is get details page!</h1>"
@app.route("/test")
def test():
  return "<h1>This is test page!</h1>"
```

```python
app.run()
```
