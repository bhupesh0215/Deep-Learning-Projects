```python
import pandas as pd
```

```python
text = "It's a pleasant          day at Bangalore; at / (10:30) am"
```

```python
series = pd.Series(text)
```

```python
series
```

```python
import texthero as hero
```

```python
hero.remove_digits(series)
```

```python
hero.remove_punctuation(series)
```

```python
hero.remove_brackets(series)
```

```python
hero.remove_whitespace(series)
```

```python
hero.remove_stopwords(series)
```

```python
hero.clean(series)
```

```python
df = pd.read_csv('https://github.com/jbesomi/texthero/raw/master/dataset/bbcsport.csv')
```

```python
df.head()
```

```python
df['pca'] = (df['text'].pipe(hero.clean).pipe(hero.tfidf).pipe(hero.pca))
```

```python
hero.scatterplot(df, 'pca', color='topic', title="PCA BBC Sport")
```

```python
df = pd.read_csv('https://github.com/jbesomi/texthero/raw/master/dataset/bbcsport.csv')
```

```python
df.head()
```

```python
df['tfidf'] = (df['text'].pipe(hero.clean).pipe(hero.tfidf))
```

```python
df['kmeans_labels'] = (df['tfidf'].pipe(hero.kmeans, n_clusters=5).astype(str))
```

```python
df['pca'] = df['tfidf'].pipe(hero.pca)
```

```python
hero.scatterplot(df,'pca', color='kmeans_labels', title='K-Means Labels')
```
