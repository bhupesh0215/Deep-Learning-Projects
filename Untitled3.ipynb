```
const_15 = tf.constant(15.0, name='constant_15.0')
const_11 = tf.constant(11.0, name='constant_11.0')
const_60 = tf.constant(60.0, name='constant_60.0')
```

```
addition = tf.add(const_11, const_15, name='adding_constant')
sustraction = tf.subtract(const_15,const_11, name='substract_constant')
multiplication = tf.multiply(const_15, const_11, name='multiplying_constant')
division = tf.divide(const_60,const_15, name='division_constant')
```

```
total_sum = tf.add_n([addition,sustraction,multiplication,division], name='tot')
```

```
print('Total sum : ', total_sum)
```
