```
import matplotlib.pyplot as plt
import pandas as pd
```

```
data = pd.read_csv('./drive/My Drive/AI/ML/Datasets/student/student-mat.csv', delimiter=';')
```

```
# data.columns.values.tolist()
```

```
# data.columns.values
# data.columns.values.tolist()
# data[['G1', 'G2', 'G3']].describe()

columns = data.columns.values.tolist()
features = columns[:-3]
labels = columns[-3:]
features, labels
```

```
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import neighbors, metrics
from sklearn.preprocessing import LabelEncoder
```

```
X = data[features].values
y = data[labels].values
```

```
# X
```

```
# y
```

```
# X[:,1]
le = LabelEncoder()
for i in range(len(features)):
  X[:,i] = le.fit_transform(X[:,i])

```

```
# plt.size()
plt.style.use('ggplot')
plt.scatter(X[:,0],y[:,0])
```

```
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```
y_train_G1 = y_train[:,0]
y_train_G2 = y_train[:,1]
y_train_G3 = y_train[:,2]

y_test_G1 = y_test[:,0]
y_test_G2 = y_test[:,1]
y_test_G3 = y_test[:,2]
```

```

X.shape, y.shape, X_train.shape, y_train.shape, X_test.shape, y_test.shape, y_train_G1.shape, y_test_G1.shape
```

```
n_neighbors = 111
knn_g1 = neighbors.KNeighborsClassifier(n_neighbors=n_neighbors, weights='uniform')
knn_g2 = neighbors.KNeighborsClassifier(n_neighbors=n_neighbors, weights='uniform')
knn_g3 = neighbors.KNeighborsClassifier(n_neighbors=n_neighbors, weights='uniform')

knn_g1.fit(X_train, y_train_G1)
knn_g2.fit(X_train, y_train_G2)
knn_g3.fit(X_train, y_train_G3)
```

```
prediction_g1 = knn_g1.predict(X_test)
prediction_g2 = knn_g2.predict(X_test)
prediction_g3 = knn_g3.predict(X_test)
```

```
accuracy_g1 = metrics.accuracy_score(y_test_G1,prediction_g1)
accuracy_g2 = metrics.accuracy_score(y_test_G2,prediction_g2)
accuracy_g3 = metrics.accuracy_score(y_test_G3,prediction_g3)
accuracy_g1, accuracy_g2, accuracy_g3
```
