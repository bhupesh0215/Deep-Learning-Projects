```
def one_hot_top_x(df, variable, top_x_labels):
  for labels in top_x_labels:
    df[variable+'_'+labels] = np.where(df[variable]==labels,1,0)
  return df
```

```
df = one_hot_top_x(df,'X2',top_10)
df = one_hot_top_x(df,'X1',[x for x in df.X1.value_counts().sort_values(ascending=False).head(10).index])
df
```
