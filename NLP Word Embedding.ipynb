```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sent = ['the glass of milk', 'the glass of juice', 'the cup of tea', 'i am a good boy', 'i am a developer', 'understand meaning of words', 'your videos are good']
sent
```

```python
voc_size = 10000
```

```python
one_hot_repr = [one_hot(words,voc_size) for words in sent]
```

```python
one_hot_repr
```

```python
from tensorflow.keras.layers import Embedding
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
sent_len=8
```

```python
embed_doc = pad_sequences(one_hot_repr, padding='pre', maxlen=sent_len)
```

```python
embed_doc
```

```python
dim=10
```

```python
from tensorflow.keras.models import Sequential
```

```python
model = Sequential()
```

```python
model.add(Embedding(voc_size,dim,input_length=sent_len))
```

```python
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(embed_doc)
```
