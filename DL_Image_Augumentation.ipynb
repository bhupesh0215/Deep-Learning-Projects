```
from keras.preprocessing.image import ImageDataGenerator, img_to_array, array_to_img, load_img
```

```
datagen = ImageDataGenerator(rotation_range=40,width_shift_range=0.2,height_shift_range=0.2,shear_range=0.2,zoom_range=0.2,horizontal_flip=True,fill_mode='nearest')
```

```
img = load_img('/content/drive/MyDrive/AI/DL/Image_Augumentation/cat.jpg')
```

```
X = img_to_array(img)
X, X.shape
```

```
X = X.reshape((1,)+X.shape)
X.shape
```

```
i = 0
for batch in datagen.flow(X,batch_size=1,save_to_dir='/content/drive/MyDrive/AI/DL/Image_Augumentation/augumented_images',save_prefix='cat',save_format='jpg'):
  i = i+1
  if i > 20:
    break
```
