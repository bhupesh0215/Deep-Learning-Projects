```
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```
data = []
for i in range(20):
  data.append(i)
data.append(101)
data.append(102)
data
```

```
def detect_outlier(data):
  outliers = []
  mean = np.mean(data)
  std = np.std(data)
  threshold = 3
  for i in data:
    z_score = np.abs((i-mean)/std)
    if z_score > threshold:
      outliers.append(i)
  return outliers
```

```
print(detect_outlier(data))
```
