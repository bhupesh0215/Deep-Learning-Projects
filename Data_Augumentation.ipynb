```
from keras.preprocessing.image import ImageDataGenerator, img_to_array, load_img
```

```
data_augument = ImageDataGenerator(rotation_range=40,width_shift_range=0.2, height_shift_range=0.2,shear_range=0.2,zoom_range=0.2,horizontal_flip=True,fill_mode='nearest')
```

```
single_img = load_img('/content/drive/MyDrive/AI/DL/FFMPEG/img-1.jpg')
```

```
image_array = img_to_array(single_img)
```

```
(1,) + image_array.shape 
```

```
image_array
```

```
image_array = image_array.reshape((1,)+image_array.shape)
```

```
image_array
```

```
# /content/drive/MyDrive/AI/DL/FFMPEG/Augymented_Image
```

```
i = 0
```

```
for batch in data_augument.flow(image_array, save_to_dir='/content/drive/MyDrive/AI/DL/FFMPEG/Augymented_Image', save_prefix='cofee', save_format='jpg'):
  i+=1
  if i > 20:
    break
```
