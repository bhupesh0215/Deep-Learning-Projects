```
black_image = np.zeros((1024,1024,3),np.uint8)
cv2_imshow(black_image)
```

```
cv2.line(black_image,(1,1),(255,255),(100,255,0),4)
cv2_imshow(black_image)
```

```
cv2.circle(black_image,(700,700),75,(255,0,0),-1)
```

```
cv2_imshow(black_image)
```

```
cv2.rectangle(black_image,(400,400),(250,250),(250,250,250),-1)
cv2_imshow(black_image)
```

```
black_image = np.zeros((720,720,3),np.uint8)
cv2.putText(black_image,'This is awesome!',(10,100),cv2.FONT_HERSHEY_SCRIPT_SIMPLEX,2,(250,170,250),6)
cv2_imshow(black_image)
```

```
black_image = np.zeros((720,720,3),np.uint8)
pts = np.array([[25,65],[415,65],[80,150],[400,500]],np.int32)
pts = pts.reshape(-1,1,2)
```

```
cv2.polylines(black_image,[pts],True,(255,255,255),3)
cv2_imshow(black_image)
```
