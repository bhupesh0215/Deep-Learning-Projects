```
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```
tf.keras.activations.deserialize('linear')
```

```
tf.keras.activations.deserialize('sigmoid')
```

```
model = keras.Sequential([layers.Dense(2,activation='relu',name='layer1'),layers.Dense(3,activation='relu',name='layer2'),layers.Dense(4,activation='relu',name='layer3')])
```

```
x = tf.ones((3,3))
x
```

```
y = model(x)
y
```

```
layer1 = layers.Dense(2,activation='relu',name='layer1')
layer2 = layers.Dense(3,activation='relu',name='layer2')
layer3 = layers.Dense(4,activation='relu',name='layer3')

y = layer3(layer2(layer1(x)))
y
```
