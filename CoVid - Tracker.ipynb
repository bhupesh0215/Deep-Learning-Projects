```python
import pandas as pd
```

```python
df = pd.read_json('https://www.mohfw.gov.in/data/datanew.json')
```

```python
df.head()
```

```python
df = df.iloc[:-1,1:]
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
total_states = np.arange(len(df['state_name']))
```

```python
total_states
```

```python
plt.figure(num=None, figsize=(9,6), dpi=80, facecolor='w', edgecolor='k')
plt.barh(total_states,df['positive'], align='center', alpha=0.5, color=(1,0,0), edgecolor=(0.5,0.2,0.8))
plt.yticks(total_states,df['state_name'])
plt.xlim(1,max(df['positive'])+100)
plt.xlabel('Positive cases')
plt.title('Corona cases')
plt.show()
```

```python
plt.figure(num=None, figsize=(9,6), dpi=80, facecolor='w', edgecolor='k')
plt.barh(total_states,df['active'], align='center', alpha=0.5, color=(1,0,0), edgecolor=(0.5,0.2,0.8))
plt.yticks(total_states,df['state_name'])
plt.xlim(1,max(df['active'])+100)
plt.xlabel('Active cases')
plt.title('Corona cases')
plt.show()
```

```python
plt.figure(num=None, figsize=(9,6), dpi=80, facecolor='w', edgecolor='k')
plt.barh(total_states,df['new_active'], align='center', alpha=0.5, color=(1,0,0), edgecolor=(0.5,0.2,0.8))
plt.yticks(total_states,df['state_name'])
plt.xlim(1,max(df['new_active'])+100)
plt.xlabel('New Active cases')
plt.title('Corona cases')
plt.show()
```

```python
plt.figure(num=None, figsize=(9,6), dpi=80, facecolor='w', edgecolor='k')
plt.barh(total_states,df['new_positive'], align='center', alpha=0.5, color=(1,0,0), edgecolor=(0.5,0.2,0.8))
plt.yticks(total_states,df['state_name'])
plt.xlim(1,max(df['new_positive'])+100)
plt.xlabel('New Positive cases')
plt.title('Corona cases')
plt.show()
```

```python
plt.figure(num=None, figsize=(9,6), dpi=80, facecolor='w', edgecolor='k')
plt.barh(total_states,df['death'], align='center', alpha=0.5, color=(1,0,0), edgecolor=(0.5,0.2,0.8))
plt.yticks(total_states,df['state_name'])
plt.xlim(1,max(df['death'])+100)
plt.xlabel('Deaths')
plt.title('Corona cases')
plt.show()
```
