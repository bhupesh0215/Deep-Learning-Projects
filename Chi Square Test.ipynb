```python
import seaborn as sns
import numpy as np
import scipy.stats as stats
```

```python
import pandas as pd
```

```python
dataset = sns.load_dataset('tips')
```

```python
dataset.head()
```

```python
dataset_table = pd.crosstab(dataset['sex'],dataset['smoker'])
```

```python
dataset_table
```

```python
obs_val = dataset_table.values
```

```python
obs_val
```

```python
val = stats.chi2_contingency(dataset_table)
```

```python
val
```

```python
expected_value = val[3]
```

```python
expected_value
```

```python
num_of_rows = dataset_table.shape[0]
```

```python
num_of_rows
```

```python
num_of_cols = dataset_table.shape[1]
```

```python
num_of_cols
```

```python
dof = (num_of_rows-1)*(num_of_cols-1)
```

```python
dof
```

```python
alpha = 0.05
```

```python
from scipy.stats import chi2
```

```python
[(o,e) for o,e in zip(obs_val,expected_value)]
```

```python
chi_sqre = sum([(o-e)**2./e for o,e in zip(obs_val,expected_value)])
```

```python
chi_sqre
```

```python
chi_sqre_stat = chi_sqre[0] + chi_sqre[1]
```

```python
chi_sqre_stat
```

```python
critical_val = chi2.ppf(q=1-alpha, df=dof)
```

```python
critical_val
```

```python
p_val  = 1 - chi2.cdf(x=chi_sqre_stat, df=dof)
```

```python
p_val
```

```python
if p_val <= alpha:
  print('Reject H0, There is a relationship between two categorical variable')
else:
  print('Retain H0, There is no relationship between two categorical variable')
```
