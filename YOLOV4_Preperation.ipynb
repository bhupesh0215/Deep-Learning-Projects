```
import os
import pandas as pd
```

```
os.chdir('/content/drive/MyDrive/AI/DL/YOLOV4/Tea_Coffee/csv_folder')
```

```
classes_data = pd.read_csv('class-descriptions-boxable.csv', header=None )
```

```
classes_data.head()
```

```
classes = ['Coffee','Tea']
```

```
classes_string = [] 
```

```
classes_data.loc[classes_data[1] == 'Coffee']
```

```
classes_data.loc[classes_data[1] == 'Tea']
```

```
for i in classes:
  req_classes = classes_data.loc[classes_data[1] == i]
  print('req_classes:' , req_classes)
  string = req_classes.iloc[0][0]
  print('string:',string)
  classes_string.append(string)
  # print(req_classes)
# classes_string
```

```
annotation_data = pd.read_csv('train-annotations-bbox.csv', usecols=['ImageID', 'LabelName','XMin','XMax','YMin','YMax'])
annotation_data
```

```
annotation_data
```

```
filtered_class_data = annotation_data.loc[annotation_data['LabelName'].isin(classes_string)].copy()
filtered_class_data
```

```
filtered_class_data['classnumber'] = ""
filtered_class_data['center_x'] = ""
filtered_class_data['center_y'] = ""
filtered_class_data['height'] = ""
filtered_class_data['weight'] = ""
```

```
for i in range(len(classes_string)):
  filtered_class_data.loc[filtered_class_data['LabelName']==classes_string[i],'classnumber'] = i
```

```
filtered_class_data
```

```
filtered_class_data['center_x'] = (filtered_class_data['XMax']+filtered_class_data['XMin'])/2
filtered_class_data['center_y'] = (filtered_class_data['YMax']+filtered_class_data['YMin'])/2
filtered_class_data
```

```
filtered_class_data['height'] = filtered_class_data['YMax'] - filtered_class_data['YMin']
filtered_class_data['weight'] = filtered_class_data['XMax'] - filtered_class_data['XMin']
filtered_class_data
```

```
YOLO_values = filtered_class_data.loc[:,['ImageID', 'classnumber', 'center_x','center_y', 'height' ,'weight']].copy()
YOLO_values
```

```
img_path = '/content/drive/MyDrive/AI/DL/YOLOV4/Tea_Coffee/Dataset/train/Coffee_Tea'
os.chdir(img_path)
```

```
for cur_dir, dirs, files in os.walk('.'):
  # print(cur_dir, dirs, files)
  for f in files:
    if f.endswith('jpg'):
      file_title = f[:-4]
      # print(file_title)
      YOLO_file = YOLO_values.loc[YOLO_values['ImageID'] == file_title]
      df = YOLO_file.loc[:,['classnumber','center_x','center_y','height','weight']]
      save_path = img_path + '/' + file_title + '.txt'
      df.to_csv(save_path,header=False,index=False,sep=' ')
# YOLO_file
```

```
path_list = []
for cur_dir, dirs, files in os.walk('.'):
  for f in files:
    if f.endswith('jpg'):
      file_name = img_path + '/' + f
      path_list.append(file_name + '\n')
path_list
```

```
path_list_test = path_list[:int(len(path_list)*0.2)]
path_list = path_list[int(len(path_list)*0.2):]
```

```
len(path_list), len(path_list_test)
```

```
with open('train.txt','w') as train:
  for i in path_list:
    train.write(i)

with open('test.txt','w') as test:
  for i in path_list_test:
    test.write(i)
```

```
i=0
with open('classes.names', 'w') as cls:
  with open('classes.txt','r') as text:
    for l in text:
      cls.write(l)
      i+=1
i
```
